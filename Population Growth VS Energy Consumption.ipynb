```python
#Dropping the irrelevant columns

df_population_growth =  df_population_growth.drop(['Country Code','Indicator Name','Indicator Code'], axis=1)
df_power_consumption =  df_power_consumption.drop(['Country Code','Indicator Name','Indicator Code'], axis=1)
```

### Dealing with missing values

```python
# To display all the values in the dataframe for our analysis
pd.set_option('display.max_rows', None)
```

```python
mv_population_growth = np.array(df_population_growth.isnull().sum())
mv_power_consumption = np.array(df_power_consumption.isnull().sum())
```

```python
mv_pg_summary = pd.DataFrame(df_population_growth.columns.values, columns=['Column Names'])
mv_pg_summary['No of Missing Values'] = mv_population_growth
mv_pg_summary = mv_pg_summary.set_index('Column Names')
mv_pg_summary.sort_values(by='No of Missing Values', ascending=False)
```

```python
mv_pc_summary = pd.DataFrame(df_power_consumption.columns.values, columns=['Column Names'])
mv_pc_summary['No of Missing Values'] = mv_power_consumption
mv_pc_summary = mv_pc_summary.set_index('Column Names')
mv_pc_summary.sort_values(by='No of Missing Values', ascending=False)
```

#### Based on the number of missing values over the years in the Power Cosumption dataframe, our analysis would be from year 1990 to 2014

```python
colnames = ['Country Name', '1990', '1991','1992', '1993', '1994', '1995', '1996', 
       '1997', '1998', '1999', '2000', '2001', '2002', '2003', '2004', 
       '2005', '2006', '2007', '2008', '2009', '2010', '2011', '2012', 
       '2013', '2014']


df_power_consumption = df_power_consumption[colnames]
df_population_growth = df_population_growth[colnames]
```

#### Using mean imputation techinque to handle missing data

```python
#from sklearn.impute import SimpleImputer

#mean_imputer = SimpleImputer(strategy = 'mean')

#df_power_consumption.iloc[:,1:] = mean_imputer.fit_transform(df_power_consumption.iloc[:,1:])
#df_population_growth.iloc[:,1:] = mean_imputer.fit_transform(df_population_growth.iloc[:,1:])
```

### Setting Missing values to zero

```python
df_power_consumption_nmv = df_power_consumption.fillna(0)
df_population_growth_nmv = df_population_growth.fillna(0)
```

```python
df_power_consumption_nmv.isnull().sum()
df_population_growth_nmv.isnull().sum()
```

```python
# Deleting countries with zero record

df_power_consumption_nmv['Country Total'] = df_power_consumption_nmv.iloc[:,1:].sum(axis = 1)
df_population_growth_nmv['Country Total'] = df_population_growth_nmv.iloc[:,1:].sum(axis = 1)

df_power_consumption_nmv = df_power_consumption_nmv[~(df_power_consumption_nmv['Country Total'] < 1)]
#df_population_growth_nmv = df_population_growth_nmv[~(df_power_consumption_nmv['Country Total'] < 1)]
```

### Saving the Top and Last 5 for each dataset

```python
# Saving the Top and Last 5 for each dataset

sorted_df_power_consumption = df_power_consumption_nmv.sort_values(by = 'Country Total', ascending = False)
sorted_df_population_growth = df_population_growth_nmv.sort_values(by = 'Country Total', ascending = False)

# Power Consumption Category
Top5_df_power_consumption = sorted_df_power_consumption[0:5]
Top5_df_power_consumption.set_index('Country Name', inplace=True)
Last5_df_power_consumption = sorted_df_power_consumption.tail(5)
Last5_df_power_consumption.set_index('Country Name', inplace=True)

# Popultion Category
Top5_df_population_growth = sorted_df_population_growth[0:5]
Top5_df_population_growth.set_index('Country Name', inplace=True)
Last5_df_population_growth = sorted_df_population_growth.tail(5)
Last5_df_population_growth.set_index('Country Name', inplace=True)
```

### Preparing the Data from Visualisation

```python
#Transposing the dataframes then save the files to CSV

Top5_df_power_consumption.T.to_csv('Top5_T_power_consumption.csv')
Top5_df_population_growth.T.to_csv('Top5_T_population_growth.csv')
Last5_df_power_consumption.T.to_csv('Last5_T_power_consumption.csv')
Last5_df_population_growth.T.to_csv('Last5_T_population_growth.csv')
```

```python
#Loading the transposed data to a new dataframe

top5_pc = pd.read_csv('Top5_T_power_consumption.csv', index_col=0)
top5_pg = pd.read_csv('Top5_T_population_growth.csv', index_col=0)
last5_pc = pd.read_csv('Last5_T_power_consumption.csv', index_col=0)
last5_pg = pd.read_csv('Last5_T_population_growth.csv', index_col=0)
```

```python
cols_Top5_pc = top5_pc.columns.values
cols_Top5_pg = top5_pg.columns.values
cols_Last5_pc = last5_pc.columns.values
cols_Last5_pg = last5_pg.columns.values
```

### Visualizing the data

```python
color_range_power=['v--r','v--g','v--b','v--y','v--k']
color_range_population=['gray','purple','orange','brown','cyan']
```

```python
fig, (ax1,ax2) = plt.subplots(1,2, figsize = (14,4))

for i in range(5):
    ax1.plot(top5_pc[cols_Top5_pc[i]], color_range_power[i], label = cols_Top5_pc[i])
ax1.set_xlabel('Years')
ax1.set_ylabel('Power Consumption')
ax1.legend()

for i in range(5):
    ax2.plot(top5_pg[cols_Top5_pg[i]], color_range_population[i], label = cols_Top5_pg[i])
ax2.set_xlabel('Years')
ax2.set_ylabel('Population Growth')
ax2.legend()
```
